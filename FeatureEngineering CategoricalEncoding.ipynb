```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data={'Temperature':['Hot','Cold','Very Hot','Warm','Hot','Warm','Warm','Hot','Hot','Cold'],
      'color':['Red','Yellow','Blue','Blue','Red','Yellow','Red','Yellow','Yellow','Yellow'],
      'Target':[1,1,1,0,1,0,1,0,1,1]}
df=pd.DataFrame(data,columns=['Temperature','color','Target'])
df
```

# Encoding Techniques not based on Target Variable
## One-hot Encoding

```python
import category_encoders as ce
ohec=ce.one_hot.OneHotEncoder()
df1=ohec.fit_transform(df['color'])
res=pd.concat([df,df1],axis=1)
res
```

## Label Encoding

```python
df1=df.copy()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Temp_Encoded']=le.fit_transform(df['Temperature'])
df1
```

## Ordinal Encoding

```python
df1=df.copy()
temp_dict={'Cold':1,
           'Warm':2,
           'Hot':3,
           'Very Hot':4
}
df1['Temp_Ordinal']=df1['Temperature'].map(temp_dict)
df1
```

## Hashing

```python
#!pip install category_encoders==2.0.0
```

```python
df1=df.copy()
df2=ce.HashingEncoder().fit_transform(df1['color'])
res=pd.concat([df1,df2],axis=1)
res
```

```python
df1=df.copy()
df1[['c_0','c_1']]=ce.HashingEncoder(n_components=2,hash_method='md5').fit_transform(df1['color'])
df1
```

## Binary Encoding

```python
df1=df.copy()
df1[['c_0','c_1','c_2']]=ce.BinaryEncoder().fit_transform(df1['color'])
df1
```

## Helmert Encoding

```python
df1=df.copy()
encoder=ce.HelmertEncoder()
dfh=encoder.fit_transform(df['color'])
df1=pd.concat([df1,dfh],axis=1)
df1
```

## Frequency Encoding

```python
df1=df.copy()
temp_dict=dict(df1['Temperature'].value_counts())
df1['Temp_Freq']=df1['Temperature'].map(temp_dict)
df1
```

# Encoding Techniques based on Target Variable
## Mean Encoding

```python
df1=df.copy()
mean_df=df1.groupby(['Temperature']).mean()
df1['Temp_mean']=df1.Temperature.map(mean_df['Target'])
df1
```

## Weight of Evidence (WOE) Encoding

```python
df1=df.copy()
mean_df=df1.groupby(['Temperature']).mean()
df1['pos_prob']=df1.Temperature.map(mean_df['Target'])
df1['neg_prob']=1-df1['pos_prob']
df1['neg_prob']=np.where(df1['neg_prob']==0,0.000001,df1['neg_prob'])
df1['Temp_WOE']=np.log(df1['pos_prob']/df1['neg_prob'])
df1
```

## Probability Ratio Encoding

```python
df1=df.copy()
mean_df=df1.groupby(['Temperature']).mean()
df1['pos_prob']=df1.Temperature.map(mean_df['Target'])
df1['neg_prob']=1-df1['pos_prob']
df1['neg_prob']=np.where(df1['neg_prob']==0,0.000001,df1['neg_prob'])
df1['Temp_WOE']=df1['pos_prob']/df1['neg_prob']
df1
```
