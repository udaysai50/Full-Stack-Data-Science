```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('cancer.csv')
df.head()
```

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
df.drop('id',axis=1,inplace=True)
df['diagnosis'].replace('M',0,inplace=True)
df['diagnosis'].replace('B',0,inplace=True)
df.info()
```

```python
df.head()
```

```python
X=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

## Variance Threshold
Variance Threshold is a baseline approach to select features with minimum threshold. This threshold is by default 0 and can be customised. This method outputs all the features which has variance above this threshold. This method is used in the scenarious when we are working with humoungous dataset and eliminate some features intuitively  

```python
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(0.85*(1-0.85))
sel_fet=vt.fit(X)
X[X.columns[sel_fet.get_support(indices=True)]].head()
```

###### WE can see that only 11 columns in the test set are retained meaning only these 11 columns has a variance above the threshold we mentioned (0.85 in the above code)

## Filter Methods
Filter Methods or Univariate methods are used to compare the feature and target one at a time. Several cases exists in performing the filter methods. Sklearn library offers several methods like SelectKBest, SelectPercentile, SelectFpr etc... which takes statistical test as an input and results the test statistic and p-value

#### SelectKBest
In the cancer dataset as we are dealing with classification problem and all the inputs are numerical we will perform an ANOVA test which is performed on the basis of a F-test

```python
from sklearn.feature_selection import SelectKBest,f_classif
sb=SelectKBest(f_classif,k=20)
sb.fit(X,y)
cols=sb.get_support(indices=True)
#print(cols)
X_new=X.iloc[:,cols]
print(len(X_new.columns))
print(X_new.columns)
X_new.head()
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif
sb=SelectKBest(mutual_info_classif,k=20)
sb.fit(X,y)
cols=sb.get_support(indices=True)
#print(cols)
X_new=X.iloc[:,cols]
print(len(X_new.columns))
print(X_new.columns)
X_new.head()
```

##### SelectPercentile
SelectPercentile takes one statistical test and percentile as inputs. It results the mentioned percentile of columns when compared to the original number of columns

```python
from sklearn.feature_selection import SelectPercentile,f_classif
sp=SelectPercentile(f_classif,percentile=75)
sp.fit(X,y)
cols=sp.get_support(indices=True)
#print(cols)
X_new=X.iloc[:,cols]
print(len(X_new.columns))
print(X_new.columns)
X_new.head()
```

```python
from sklearn.feature_selection import SelectPercentile,mutual_info_classif
sp=SelectPercentile(mutual_info_classif,percentile=75)
sp.fit(X,y)
cols=sp.get_support(indices=True)
#print(cols)
X_new=X.iloc[:,cols]
print(len(X_new.columns))
print(X_new.columns)
X_new.head()
```

##### SelectKBest for Categorical feature and Categorical target
In this case we implement a chi-square test which constructs a contingency table under the hood and retrieves the specified 'k' features based on the values of the Chi-statistic calculated

```python
df=pd.read_csv('https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/churn_data_st.csv')
df.head()
```

```python
X=df[['gender','PaperlessBilling']]
y=df['Churn']
y.head()
```

```python
X['gender'].replace('Female',1,inplace=True)
X['gender'].replace('Male',0,inplace=True)
X['PaperlessBilling'].replace('Yes',1,inplace=True)
X['PaperlessBilling'].replace('No',0,inplace=True)
y.replace('No',0,inplace=True)
y.replace('Yes',1,inplace=True)
```

```python
X.head()
```

```python
from sklearn.feature_selection import SelectKBest,chi2
sb=SelectKBest(chi2,k=1)
sb.fit(X,y)
cols=sb.get_support(indices=True)
X_new=X.iloc[:,cols]
print(X_new.columns)
X_new.head()
```

##### SelectPercentile for Categorical feature and Categorical target
In this case we implement a chi-square test which constructs a contingency table under the hood and retrieves the specified 'percentile' features based on the values of the Chi-statistic calculated

```python
from sklearn.feature_selection import SelectPercentile,chi2
sp=SelectPercentile(chi2,percentile=50)
sp.fit(X,y)
cols=sp.get_support(indices=True)
X_new=X.iloc[:,cols]
print(X_new.columns)
X_new.head()
```

#### SelectKBest For Regression
We use a f_regression which implements correlation under the hood

```python
from sklearn.datasets import load_boston
X,y=load_boston(return_X_y=True)
X[0:1]
```

```python
from sklearn.feature_selection import SelectKBest,f_regression
sb=SelectKBest(f_regression,k=10)
sb.fit(X,y)
cols=sb.get_support(indices=True)
print(cols)
```
